```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

def analyze_ratings(data_path):

    df = pd.read_csv(data_path, index_col=0)

    # Basic counts
    num_users = df.shape[1]
    num_items = df.shape[0]
    unique_users = df.columns.tolist()
    unique_items = df.index.tolist()

    # Ratings per item
    ratings_per_item = df.notna().sum(axis=1)

    # Distribution of ratings
    all_ratings = df.stack().astype(float)
    plt.hist(all_ratings, bins=np.arange(0.5, 6, 1), edgecolor="black")
    plt.title("Distribution of Ratings")
    plt.xlabel("Rating")
    plt.ylabel("Frequency")
    plt.show()

    # Matrix sparsity and bias
    total_possible_ratings = df.size
    total_actual_ratings = df.notna().sum().sum()
    sparsity = 1 - (total_actual_ratings / total_possible_ratings)

    mean_rating = all_ratings.mean()
    rating_counts = df.stack().value_counts()

    # Lowest-rated items
    average_ratings_per_item = df.mean(axis=1)
    lowest_rated_items = average_ratings_per_item.nsmallest(2).index.tolist()

    # Save results
    results = {
        "num_users": num_users,
        "num_items": num_items,
        "unique_users": unique_users,
        "unique_items": unique_items,
        "ratings_per_item": ratings_per_item.to_dict(),
        "matrix_sparsity": sparsity,
        "mean_rating": mean_rating,
        "rating_counts": rating_counts.to_dict(),
        "lowest_rated_items": lowest_rated_items,
    }

    return results


def main():
    # Replace with your actual data path
    data_path = "spotify_tracks_with_random_ratings.csv"
    results = analyze_ratings(data_path)

    # Print summary
    print("Summary of Results:")
    for key, value in results.items():
        print(f"{key}: {value}")


if __name__ == "__main__":
    main()
```

```python
import pandas as pd
import numpy as np

def collaborative_filtering(ratings_file, target_tracks):
    # Load the dataset
    ratings_df = pd.read_csv(ratings_file, index_col=0)

    # Calculate average rating for target tracks
    track_avg_ratings = {}
    for track in target_tracks:
        track_avg_ratings[track] = ratings_df.loc[track].mean()

    # Impute missing ratings with track averages
    for track in target_tracks:
        ratings_df.loc[track] = ratings_df.loc[track].fillna(track_avg_ratings[track])

    # Calculate average rating for each track
    track_averages = ratings_df.mean(axis=1)

    # Calculate rating deviations from track averages
    rating_deviations = ratings_df.sub(track_averages, axis=0)

    # Compute covariance matrix of rating deviations
    track_covariance_matrix = rating_deviations.T.cov()

    # Determine top 5 and top 10 peers for target tracks
    top_peers = {}
    for track in target_tracks:
        track_covariances = track_covariance_matrix.loc[track]
        sorted_tracks = track_covariances.sort_values(ascending=False).index
        top_peers[track] = {
            "top_5": sorted_tracks[1:6].tolist(),  # Exclude the track itself
            "top_10": sorted_tracks[1:11].tolist(),
        }

    # Create reduced dimensional spaces (top 5 and top 10 peers)
    ratings_transposed = ratings_df.T  # Transpose to user-track format
    reduced_space_top_5 = {}
    reduced_space_top_10 = {}
    for track in target_tracks:
        peers_5 = top_peers[track]["top_5"]
        peers_10 = top_peers[track]["top_10"]
        reduced_space_top_5[track] = ratings_transposed[peers_5].fillna(0)
        reduced_space_top_10[track] = ratings_transposed[peers_10].fillna(0)

    # Predict missing ratings using top 5 and top 10 peers
    predicted_ratings_top_5 = {}
    predicted_ratings_top_10 = {}
    for track in target_tracks:
        predicted_ratings_top_5[track] = reduced_space_top_5[track].mean(axis=1)
        predicted_ratings_top_10[track] = reduced_space_top_10[track].mean(axis=1)

    return {
        'track_avg_ratings': track_avg_ratings,
        'top_peers': top_peers,
        'predicted_ratings_top_5': predicted_ratings_top_5,
        'predicted_ratings_top_10': predicted_ratings_top_10
    }

# Example usage
if __name__ == "__main__":
    ratings_file = "spotify_tracks_with_random_ratings.csv"
    target_tracks = ['If It Kills Me', 'Ghost - Acoustic']
    results = collaborative_filtering(ratings_file, target_tracks)

    # Print results (example)
    print("\nPredicted Ratings (Top 5 Peers):")
    for track, predictions in results['predicted_ratings_top_5'].items():
        print(f"\nPredicted ratings for {track}:\n")
        print(predictions.to_frame(name="Predicted Rating").head())
```

```python
import pandas as pd
import numpy as np

def collaborative_filtering_mle(ratings_file, target_tracks):
    # Load the dataset
    ratings_df = pd.read_csv(ratings_file, index_col=0)

    # Calculate average rating for target tracks
    track_avg_ratings = {}
    for track in target_tracks:
        track_avg_ratings[track] = ratings_df.loc[track].mean()

    # Impute missing ratings with track averages
    for track in target_tracks:
        ratings_df.loc[track] = ratings_df.loc[track].fillna(track_avg_ratings[track])

    # Calculate average rating for each track
    track_averages = ratings_df.mean(axis=1)

    # Calculate rating deviations from track averages
    rating_deviations = ratings_df.sub(track_averages, axis=0)

    # Compute covariance matrix using MLE
    def mle_covariance_matrix(ratings_df):
        """Calculate the covariance matrix using MLE."""
        item_names = ratings_df.index
        cov_matrix = pd.DataFrame(index=item_names, columns=item_names, dtype=float)

        for item1 in item_names:
            for item2 in item_names:
                common_users = ratings_df.loc[[item1, item2]].dropna(axis=1, how='any')
                if len(common_users.columns) > 0:
                    cov_matrix.loc[item1, item2] = np.cov(common_users.loc[item1], common_users.loc[item2], bias=True)[0, 1]
                else:
                    cov_matrix.loc[item1, item2] = 0

        return cov_matrix

    mle_cov_matrix = mle_covariance_matrix(ratings_df)

    # Determine top 5 and top 10 peers for target tracks using MLE covariance
    top_peers_mle = {}
    for track in target_tracks:
        track_covariances = mle_cov_matrix.loc[track]
        sorted_tracks = track_covariances.sort_values(ascending=False).index
        top_peers_mle[track] = {
            "top_5": sorted_tracks[1:6].tolist(),  # Exclude the track itself
            "top_10": sorted_tracks[1:11].tolist(),
        }

    # Create reduced dimensional spaces (top 5 and top 10 peers)
    ratings_transposed = ratings_df.T  # Transpose to user-track format
    reduced_space_mle_top_5 = {}
    reduced_space_mle_top_10 = {}
    for track in target_tracks:
        peers_5 = top_peers_mle[track]["top_5"]
        peers_10 = top_peers_mle[track]["top_10"]
        reduced_space_mle_top_5[track] = ratings_transposed[peers_5].fillna(0)
        reduced_space_mle_top_10[track] = ratings_transposed[peers_10].fillna(0)

    # Predict missing ratings using top 5 and top 10 peers (MLE)
    predicted_ratings_mle_top_5 = {}
    predicted_ratings_mle_top_10 = {}
    for track in target_tracks:
        predicted_ratings_mle_top_5[track] = reduced_space_mle_top_5[track].mean(axis=1)
        predicted_ratings_mle_top_10[track] = reduced_space_mle_top_10[track].mean(axis=1)

    return {
        'track_avg_ratings': track_avg_ratings,
        'top_peers_mle': top_peers_mle,
        'predicted_ratings_mle_top_5': predicted_ratings_mle_top_5,
        'predicted_ratings_mle_top_10': predicted_ratings_mle_top_10
    }

# Example usage
if __name__ == "__main__":
    ratings_file = "spotify_tracks_with_random_ratings.csv"
    target_tracks = ['If It Kills Me', 'Ghost - Acoustic']
    mle_results = collaborative_filtering_mle(ratings_file, target_tracks)

    # Print results (example)
    print("\nPredicted Ratings (MLE Top 5 Peers):")
    for track, predictions in mle_results['predicted_ratings_mle_top_5'].items():
        print(f"\nPredicted ratings for {track} (Top 5-Peers):\n")
        print(predictions.to_frame(name="Predicted Rating").head())
```
